```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
Sex: male = 1; female = 0
cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)
trestbps: The person's resting blood pressure (mm Hg on admission to the hospital)
chol: The person's choesterol measurement in mg/dl
fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)
restecg: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)
thalach: The person's maximum heart rate achieved
exang: Exercise induced angina (1 = yes; 0 = no)
oldpeak: ST depression induced by exercise relative to rest ('ST' relates to positions on the ECG plot. See more here)
slope: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)
ca: The number of major vessels (0-3)
thal: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect)
target: Heart disease (0 = no, 1 = yes)
"""

data = pd.read_csv('data/heart.csv')
data.head()
#data.columns
```

```python
data['sex'] = data['sex'].replace([0,1],["Female","Male"])
```

```python
fig,ax = plt.subplots()
 
data[data.sex == "Female"].thal.value_counts().plot(kind = "bar", figsize=(5,5), title = "Heart disease: Chest pain type")

plt.xlabel("Female")
plt.ylabel("Thalassemia")
plt.show()

data[data.sex == "Male"].thal.value_counts().plot(kind = "bar", figsize=(5,5), title = "Heart disease: Chest pain type")

plt.xlabel("Male")
plt.ylabel("Thalassemia")
plt.show()
```

```python
fig = plt.figure(figsize=(15,15))
corr = data.corr()
sns.heatmap(corr, annot=True, vmin=-1.0,vmax=1.0)
```

```python
fig = plt.figure(figsize=(15,10))
sns.set_style("white")
ax=sns.distplot(data['trestbps'],axlabel="Patient Resting Blood Pressure", bins=20, kde=True)
sns.set_context("notebook")
ax.set_title("Histogram of Resting Blood Pressure", fontsize=25)
ax.set_xlabel("Resting Blood Pressure",fontsize=15)
ax.set_ylabel("Frequency",fontsize=15)
ax.tick_params(labelsize=15)
```

```python
fig = plt.figure(figsize=(10,10))
ax = sns.boxplot(x='cp', y='trestbps', hue='sex', data=data)
ax.set_title("Boxplot of Chest Pain and Resting Blood Pressure", fontsize=25)
ax.set_xlabel("Chest Pain Type",fontsize=15)
ax.set_ylabel("Resting Blood Pressure",fontsize=15)
ax.tick_params(labelsize=15)
```

```python
fig = plt.figure(figsize=(15,15))

ax = sns.catplot(x='cp', y='trestbps', hue="sex", kind="swarm", data=data)
plt.title("Swarmplot of Chest Pain and Resting Blood Pressure By Sex", fontsize=18)
plt.xlabel("Chest Pain",fontsize=15)
plt.ylabel("Resting Blood Pressure",fontsize=15)
plt.tick_params(labelsize=15)
```
