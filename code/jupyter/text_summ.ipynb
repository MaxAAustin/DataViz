```python
import PyPDF2
import bs4 
import urllib.request as req
import re
import nltk
import heapq
```

```python
file = open('qc.pdf', 'rb')

pdf_reader = PyPDF2.PdfFileReader(file)

print(pdf_reader.numPages)
```

```python
page_obj = pdf_reader.getPage(0)
print(page_obj.extractText())

text = page_obj.extractText()

file.close()
```

```python
text = re.sub(r'\[[0-9]*\]',' ', text)
text = re.sub(r'\s+', ' ', text)


format_text = re.sub('[^a-zA-Z]',' ', text)
format_text = re.sub(r'\s+',' ', format_text)


sent_list = nltk.sent_tokenize(text)

stopwords = nltk.corpus.stopwords.words('english')

art_tokens = nltk.word_tokenize(format_text)

word_freqs = {}

for word in art_tokens:
    if word not in stopwords:
        if word not in word_freqs.keys():
            word_freqs[word] = 1
        else:
            word_freqs[word] += 1


max_freq = max(word_freqs.values())

for word in word_freqs.keys():
    word_freqs[word] = (word_freqs[word]/max_freq)
    
sent_scores = {}

for sent in sent_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_freqs.keys():
            if len(sent.split(' ')) < 30:
                if sent not in sent_scores.keys():
                    sent_scores[sent] = word_freqs[word]
                else:
                    sent_scores[sent] += word_freqs[word]
                    
summ_sents = heapq.nlargest(7, sent_scores, key=sent_scores.get)

summary = ' '.join(summ_sents)

print(summary,"\n")
for s in summ_sents:
    print(s, "\n")

```
