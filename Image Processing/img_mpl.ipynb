```python
"""
Image Processing: Matplotlib Version
Imports: Matplotlib and NumPy. Numpy also used in OpenCV version because of efficient memory usage and pixels stored
as arrays of values. 
"""

from matplotlib import pyplot as plt
import matplotlib.image as mpi
from matplotlib.gridspec import GridSpec as gs
import numpy as np
import PIL.Image as im 
import PIL.ImageEnhance as ie
```

```python
#img = cv2.imread('th.jpeg',0)

#cv2.imshow("Hardy", img)
plt.figure(figsize=(15,10))

"""
Will be used as our color map variable
"""
c = 'gray'

img = mpi.imread('th.jpeg', format='JPEG')


"""
Gives image data of the red, green, and blue layers, respectively.
"""
img_red = img[:,:,0]
img_green = img[:,:,1]
img_blue = img[:,:,2]

"""
Removes the x,y axis ticks and labels
"""
plt.axis('off')
plt.title('Commander of the Most Excellent Order of the British Empire Tom Hardy', fontsize=24)

iplot = plt.imshow(img_green, interpolation='nearest')
```

```python
fig = plt.figure(figsize=(20,20))

grid = gs(1,3)

r = fig.add_subplot(231)
r.imshow(img_red, interpolation='nearest')
r.axis('off')

g = fig.add_subplot(232)
g.imshow(img_green, interpolation='nearest')
g.axis('off')

b = fig.add_subplot(233)
b.imshow(img_blue, interpolation='nearest')
b.axis('off')
```
